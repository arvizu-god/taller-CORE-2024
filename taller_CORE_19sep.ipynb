## Quantum Fourier Transform: PennyLane

### PennyLane

La **Transformada Cuántica de Fourier (QFT)** es una versión cuántica de la Transformada de Fourier clásica, una herramienta matemática que nos permite descomponer una función o señal en sus componentes de frecuencia.

En el contexto de la computación cuántica, la QFT se aplica a los estados cuánticos y es esencial en muchos algoritmos cuánticos importantes, como el algoritmo de Shor para la factorización de números grandes.

**Conceptos clave:**


*   **Transformada de Fourier clásica:** Convierte una señal del dominio del tiempo al dominio de la frecuencia. En el caso de la QFT, hacemos algo similar pero con amplitudes de probabilidad en estados cuánticos.
*  **Superposición cuántica:** La QFT aprovecha la superposición de estados cuánticos para procesar de forma eficiente toda la información contenida en un sistema cuántico. Esto es lo que permite que la QFT sea exponencialmente más rápida que su contraparte clásica.
*   **Operación eficiente:** Mientras que una transformada de Fourier clásica puede tomar $O(n 2^n)$ operaciones, la QFT lo hace en solo$O(n^2)$ (siendo n el número de qubits), lo que es una mejora significativa. Esta propiedad es clave para la velocidad de los algoritmos cuánticos.
*   **Aplicación:** Una de las aplicaciones más conocidas de la QFT es en el algoritmo de Shor, que permite factorizar grandes números enteros mucho más rápido que los algoritmos clásicos. Este es uno de los ejemplos que muestran cómo la computación cuántica puede resolver problemas que son inabordables con las computadoras clásicas.


**Ejemplo de la mejora significativa**
Imagina que tienes un sistema cuántico con 10 qubits. Un sistema clásico tendría que realizar $O( 2^{10})$ operaciones para realizar una Transformada de Fourier completa sobre todos los posibles estados de ese sistema (es decir, 1024 operaciones). En cambio, la QFT solo necesitaría alrededor de $O(10^2)$ operaciones, es decir, 100 operaciones. Esta mejora se vuelve aún más impresionante conforme el número de qubits aumenta.







## Draper Adder: PennyLane

### PennyLane

**1. El problema del sumador clásico**

En una computadora clásica, para sumar dos números de n bits, el proceso implica un algoritmo de suma binaria que recorre cada bit, manejando acarreos (cuando la suma de dos bits genera un "1" que debe sumarse al siguiente dígito). La suma clásica requiere un número de operaciones proporcional a O(n), ya que debe sumar bit a bit desde el menos significativo hasta el más significativo. Este proceso es secuencial, y los acarreos (carry) pueden propagarse a lo largo de toda la longitud de los números.

**2. Sumador cuántico basado en QFT (algoritmo de Draper)**

El sumador cuántico utiliza una técnica mucho más eficiente que el método clásico, ya que se basa en la Transformada Cuántica de Fourier para realizar la suma en el dominio de la fase de los qubits. El algoritmo de Draper implementa este enfoque.

**¿Cómo funciona?**

**Representación binaria:** En lugar de trabajar directamente con los números en su representación binaria estándar, el algoritmo transforma los números a una representación frecuencial usando la QFT. En esta representación, las operaciones de suma se traducen en desplazamientos de fase.

**Transformada Cuántica de Fourier:**
La clave del sumador cuántico es que convierte los números de su representación en bits a una superposición cuántica que codifica las frecuencias de estos números en las fases de los qubits.

Después de aplicar la QFT al número que queremos sumar, la suma de los números se reduce a sumar las fases en el dominio de Fourier. Como la QFT convierte la suma en un problema de fases, las operaciones se pueden realizar con desplazamientos de fase controlados, que son mucho más eficientes.

**Operación de suma:** Al realizar un desplazamiento de fase adecuado a cada qubit (en lugar de un acarreo bit a bit), se realiza la operación de suma. Esto evita el proceso secuencial de manejar acarreos que requiere una suma clásica.

**Transformada Inversa:** Una vez realizada la suma en el dominio de Fourier (mediante desplazamientos de fase), se aplica la Transformada Cuántica de Fourier Inversa (IQFT) para volver a convertir el número a su representación binaria estándar.

**Ventajas clave del uso de QFT en el sumador cuántico:**
Mayor eficiencia: El uso de la QFT reduce la complejidad del algoritmo cuántico. Mientras que un sumador clásico tiene una complejidad de O(n) (donde n es el número de bits), el sumador cuántico basado en QFT realiza la suma en O(n²), lo cual sigue siendo eficiente, considerando que el número de operaciones puede ser mucho menor cuando estamos sumando números representados en muchos qubits.

**Suma paralelizada:** La suma en el dominio de Fourier se realiza en paralelo. En lugar de recorrer bit a bit como en una suma clásica, los qubits en superposición permiten que los números se procesen de manera simultánea y más rápida, evitando el problema de los acarreos.

**3. Relación con el Algoritmo de Shor y otros Algoritmos Cuánticos**
El sumador cuántico usando QFT tiene aplicaciones más allá de simplemente sumar números. Es un componente esencial de algoritmos más complejos, como el algoritmo de Shor para la factorización de números grandes, donde es necesario realizar operaciones aritméticas rápidamente dentro de un circuito cuántico. Como el algoritmo de Shor depende de la aritmética modular (realizar operaciones matemáticas en grandes números), el sumador cuántico se usa para hacer estas sumas de manera eficiente en el contexto cuántico.


```python
#Librerias
import pennylane as qml
import numpy as np
```

```python
# Número de qubits o "wires" que usaremos en el dispositivo cuántico.
# Cada wire representa un qubit que se usará para la computación cuántica.
n_wires = 4

# Definimos un dispositivo cuántico simulado usando Pennylane. En este caso,
# usamos el dispositivo 'default.qubit', que es un simulador clásico de qubits.
# 'wires' define cuántos qubits se usarán, y 'shots=1' indica que tomaremos una muestra por ejecución.
dev = qml.device("default.qubit", wires=n_wires, shots=1)

# Esta función realiza un desplazamiento de fase controlado (rotación en Z) en cada qubit
# en el dominio de Fourier. Aquí es donde se ajustan las fases en función del número
# que estamos sumando (el valor de k).
def add_k_fourier(k, wires):
    # Iteramos sobre cada qubit (wire) para aplicar la rotación en Z.
    for j in range(len(wires)):
        # Aplicamos una rotación en el eje Z (RZ) controlada por el valor de k.
        # El ángulo de rotación es k * π / 2^j, donde j es el índice del qubit.
        # Esto implementa el desplazamiento de fase correspondiente.
        qml.RZ(k * np.pi / (2**j), wires=wires[j])

# Definimos un circuito cuántico (QNode) que suma dos números.
# El circuito recibe dos entradas: el número 'm' en binario y el número 'k' a sumar.
@qml.qnode(dev)
def sum(m, k):
    # Paso 0: Codificación del número m en el estado cuántico.
    # BasisEmbedding coloca el número m en la base computacional de los qubits.
    # Cada qubit representará uno de los bits de 'm'.
    qml.BasisEmbedding(m, wires=range(n_wires))  # m encoding

    # Paso 1: Aplicar la Transformada Cuántica de Fourier (QFT) a los qubits.
    # La QFT transforma el estado de los qubits a una representación en el dominio de Fourier,
    # donde la suma de dos números es más eficiente. Esto es un paso esencial en el sumador cuántico.
    qml.QFT(wires=range(n_wires))  # step 1

    # Paso 2: Aplicar el desplazamiento de fase correspondiente al número k.
    # Aquí, 'k' es el segundo número que queremos sumar al número 'm'.
    # La función add_k_fourier aplica el desplazamiento de fase en los qubits.
    add_k_fourier(k, range(n_wires))  # step 2

    # Paso 3: Aplicar la Transformada Cuántica de Fourier Inversa (IQFT).
    # Esto devuelve el estado cuántico al dominio de la base computacional,
    # donde podremos medir los resultados de la suma.
    qml.adjoint(qml.QFT)(wires=range(n_wires))  # step 3

    # Retornar una muestra del estado de los qubits después de la operación.
    # 'qml.sample()' medirá el estado de los qubits, devolviendo el resultado en binario.
    return qml.sample()

# Aquí solicitamos al usuario que ingrese los dos números que desea sumar.
a = int(input("Introduce el primer número: "))
b = int(input("Introduce el segundo número: "))

# Convertimos los números a binario y ejecutamos el circuito cuántico para sumarlos.
# El resultado será un array binario que representa la suma de 'a' y 'b'.
binary = sum(a, b)

# Crear un array con las potencias de 2 correspondientes a los índices de los bits.
# Esto nos ayudará a convertir el número binario de vuelta a su representación decimal.
potencias_dos = 2 ** np.arange(len(binary))[::-1]

# Calcular el valor decimal de la suma usando el producto punto entre el resultado binario
# y las potencias de 2. Esto nos da el número final en base 10 (decimal).
decimal = np.dot(binary, potencias_dos)

# Imprimir los resultados. Mostramos la suma en su representación binaria (ket) y decimal.
print(f"\nLa suma de {a} y {b} es:\n")
print(f"Representación binaria:{binary}\n")
print(f"Representación decimal:{decimal}")
```

## ¿Qué es Qiskit?

Qiskit hace referencia a un conjunto de software que ejecutan algoritmos en computadoras cuánticas. [Introduction to Qiskit](https://docs.quantum.ibm.com/guides)

Modelar, simular y crear circuitos cuánticos con Qiskit es una tarea muy distinta en comparación a otras librerías de Python (Pennylane, Circ, etc) pues es un proceso que consta de varias fases y detalles que permiten personalizar una simulación al grado que uno desee (o de la manera que se busque). Existen distintas herramientas de software que nos ayudan a mapear, optimizar, ejecutar y analizar el circuito cuántico de nuestro interés.

El Qiskit SDK (Software Development Service) es un SDK que contiene herramientas de creación de circuitos, una amplia librería de operadores y herramientas cuánticas, un transpilador el cual transforma y adapta circuitos cuánticos para que se ajusten a distintas topologías de hardware y primitivas a partir de las cuales distintos proveedores de QPU's pueden derivar sus propias implementaciones.

El QiskitRunTime service, es un servicio basado en la nube el cual permite ejecutar circuitos cuánticos en hardware real.

En el ecosistema de Qiskit, existen proyectos ajenos a Qiskit pero que proveen de distintas funcionalidades como son:
*   Qiskit Aer (qiskit-aer), el cual es un paquete que permite realizar simulaciones con modelos de ruido realisticos

Existen muchos más paquetes del ecosistema de Qiskit para distintas funcionalidades. Pueden consultar una lista completa en este link: [IBM Quantum Ecosystem](https://www.ibm.com/quantum/ecosystem).

## Qiskit Development Workflow
Los patrones de Qiskit son frameworks para descomponer problemas en distintas etapas. Los grupos de etapas son las que conforman un patrón en el cual están ivokucradas distintas de los paquetes pertenecientes al Qiskit SDK o de su ecosistema. Una vez descompuesto el problema en un patrón, este puede ser un ejecutado de manera local o mediante servicios en la nube. De manera general son 4 las etapas de un patrón de Qiskit:

![alt text](https://docs.quantum.ibm.com/images/qiskit-patterns/patterns.svg)

De manera general, consiste en cuatro pasos:

*   Mapeo: Se refiere al proceso de traducir un problema clásico o un problema cuántico de alto nivel a un formato que pueda ser procesado por una computadora cuántica. Este proceso implica diseñar circuitos cuánticos o definir operadores cuánticos que representen la lógica del problema a resolver. En esta etapa, se deben considerar aspectos como la formulación matemática del problema, la selección de algoritmos cuánticos apropiados (por ejemplo, algoritmos de optimización, búsqueda, simulación de sistemas cuánticos, etc.) y la construcción de un circuito cuántico que refleje esta lógica. Los productos obtenidos en esta etapa son circuitos cuánticos o operadores, listos para ser refinados y ejecutados en un hardware cuántico o un simulador.

*   Optimización: En esta etapa, los circuitos u operadores cuánticos obtenidos del mapeo son optimizados para su ejecución eficiente en el hardware cuántico. Este proceso incluye varios pasos, como el ajuste del circuito para reducir la cantidad de puertas, minimizar la profundidad del circuito, y disminuir los errores asociados. Además, la optimización involucra la asignación de qubits físicos mediante un proceso conocido como layout, y la ruteo de qubits mediante la inserción de compuertas SWAP para cumplir con las restricciones de conectividad del hardware. También se realiza la decomposición de compuertas para traducir el circuito a las compuertas base del hardware específico (como CNOT, RX, RZ, etc.). El objetivo principal es reducir el número de operaciones y adaptarse a las características y limitaciones del hardware, mejorando la fidelidad y eficiencia del circuito.

*   Ejecución: Consiste en ejecutar los circuitos cuánticos previamente optimizados en hardware cuántico real o en simuladores cuánticos, con el fin de obtener los resultados del cálculo cuántico. En esta etapa, los circuitos optimizados se envían a un backend, que puede ser un dispositivo cuántico físico o un simulador de alta fidelidad. Para llevar a cabo la ejecución, se utilizan primitivas como Sampler y Estimator, que facilitan la obtención de resultados al gestionar las mediciones y cálculos esperados de los circuitos. El Sampler permite la recolección de muestras de las distribuciones de resultados, mientras que el Estimator evalúa los valores esperados de observables específicos. Esta etapa también incluye la recopilación y postprocesamiento de los datos medidos, que se interpretan para extraer la información final necesaria para resolver el problema planteado.

*   Post-procesamiento: La etapa final del workflow donde se enfoca en hacer análisis de los resultados obtenidos como son visualizaciones, técnicas de mitigación de error, etc.

Para más información, consulte la documentación de Qiskit en el siguiente enlace: [Introduction to Qiskit patterns](https://docs.quantum.ibm.com/guides/intro-to-patterns)

Una vez importada la librería y nuestro circuito implementado, procedemos a medir cada qubit a la salida del circuito, transpilar el circuito y simularlo. Pero antes, ¿qué es transpilar?. Transpilar hace referencia a traducir las compuertas u operadores utilizados en el circuito en compuertas base que son las manipulables en nuestro sistema de hardaware. También adapta el circuito a la topología disponible del QPU. Muchas palabras, algo más simple.
![alt text](https://docs.quantum.ibm.com/_next/image?url=%2Fimages%2Fapi%2Fqiskit%2Ftranspiling_core_steps.png&w=1920&q=75)

El proceso de transpilado consiste de tres etapas.

1. Traducción: Nosotros somos libres de escribir un circuito con las compuertas que deseemos pero en la gran mayoria de dispositivos de hardware, solamente se tiene control de un conjunto específico de compuertas (las cuales denominamos como base). Por lo que se debe reescribir cada circuito en función de dichas compuertas. Veamos con nuestro circuito de ejemplo. Si analizamos, vemos que las compuertas de nuestro ejemplo fueron descompuestas en compuertas unitarias pues estas son aquellas que pueden ser manipuladas en hardware real.

```python
from qiskit.providers.fake_provider import GenericBackendV2
backend = GenericBackendV2(2)
bell_qc.measure_all()
circuit_basis = transpile(bell_qc, backend)
circuit_basis.draw(output='mpl')
```

2. Layout: Esta etapa consiste en mapear los qubits teóricos a los qubits reales mediante una asignación uno a uno. Es verdaderamente importante analizar en qué cadena de qubits reales se implementará el algoritmo pues eso determina el numero de compuertas adicionales que se utilicen para ejecutar el algoritmo de manera correcta. Por ejemplo, debido a que no todos los qubits están conectados entre sí (siendo muy nerds, decimos que el grafo no es conexo) si deseamos implementar compuertas de dos o más qubits entre nodos que no estén uno al lado del otro, debemos implementar compuertas $SWAP$ para cambiar su orden.

![alt text](https://docs.quantum.ibm.com/_next/image?url=%2Fimages%2Fapi%2Fqiskit%2Fmapping.png&w=1920&q=75)

3. Optimización: El descomponer el circuito en compuertas bases y el añadir compuertas $SWAP$ para ajustar el circuito a la topología del hardware, termina siendo un proceso que añade más compuertas que las originales. Afortunadamente existe rutinas de optimización que ayudan a reducir en gran manera el número de compoertas extra. Y e algunos casos los circuitos resultantes, tienen un menor número de compuertas que el original.

Mas información en la documentación de qiskit [Transpiler](https://docs.quantum.ibm.com/api/qiskit/transpiler)


Una vez discutido el proceso de transpilación, pasamos al proceso de simulación. qiskit-aer es una poderosa herramienta de simulación de circuitos cuánticos diseñada para emular el comportamiento de circuitos cuánticos en una computadora clásica. Ofrece varios métodos de simulación, cada uno optimizado para diferentes tipos de circuitos y necesidades de simulación.

Contiene distintos simuladores de los cuales destacamos los siguientes:
1. statevector: Un simulador que da como resultado las mediciones de circuitos ideales al final de los mismos.
2. density_matrix: Un simulador que permite obtener mediciones de circuitos con ruido con las mediciones al final del circuito.
3. matrix_product_state: Un simulador basado en una representación matricial de los estados de un sistema.

De manera general los distintos simuladores utilizan distintas representaciones matemáticas de los estados cuánticos para poder obtener resultados.

```python
"""
En este código implementamos el proceso de transpilación y de simulación del circuito
"""

bell_qc.measure_all() #Para poder transpilar el circuito necesitamos medir los estados de los qubits
shots=10000 #Repetimos las medidones una cantidad finita de veces y graficamos los histogramas para ver las probabilidades para cada estado

# Optimización del circuito usando qiskit-aer
simulator = AerSimulator()
bell_qc = transpile(bell_qc, simulator)
result = simulator.run(bell_qc).result()
counts = result.get_counts(bell_qc)

# Simulación usando el método Statevector
sim_statevector = AerSimulator(method='statevector')
job_statevector = sim_statevector.run(bell_qc, shots=shots)
counts_statevector = job_statevector.result().get_counts(0)

# Simulación usando el método Density Matrix
sim_density = AerSimulator(method='density_matrix')
job_density = sim_density.run(bell_qc, shots=shots)
counts_density = job_density.result().get_counts(0)

# Simulación usando el método Matrix Product State
sim_mps = AerSimulator(method='matrix_product_state')
job_mps = sim_mps.run(bell_qc, shots=shots)
counts_mps = job_mps.result().get_counts(0)


plot_histogram([ counts_statevector,counts_density, counts_mps],
               title='Bell state counts for different simulation methods',
               legend=[ 'statevector',
                       'density_matrix','matrix_product_state'])
```

## Quantum Fourier Transform: IBM-Qiskit

Una vez definidas las etapas de producción de un circuito cuántico, procedemos a implementar los algoritmos vistos en clase iniciando por la transformada cuántica de Fourier: QFT.

```python
def qft_rotations(circuit, n):
    """Realiza la qft en los primeros n qubits del circuito sin swaps"""
    if n == 0:
        return circuit
    n -= 1
    circuit.h(n)
    for qubit in range(n):
        circuit.cp(pi/2**(n-qubit), qubit, n)
    qft_rotations(circuit, n)


def swap_registers(circuit, n):
    """En esta función aplicamos operadores SWAP """
    for qubit in range(n//2):
        circuit.swap(qubit, n-qubit-1)
    return circuit

def qft(circuit, n):
    """QFT on the first n qubits in circuit"""
    qft_rotations(circuit, n)
    swap_registers(circuit, n)
    return circuit
```

```python
# Número de qubits
n = 2
qft_qc = QuantumCircuit(n, n)

# Estado inicial: |1⟩ en el qubit más significativo
qft_qc.x(n-1)  # Aplicamos X al qubit de mayor índice

# Aplicamos la QFT
qft(qft_qc, n)

# Medimos todos los qubits
qft_qc.measure_all()

# Dibujamos el circuito
qft_qc.draw('mpl')
#plt.show()
```

### Draper Adder: IBM-Qiskit

```python
from qiskit.circuit.library import QFT
#from qiskit.visualization import plot_histogram, plot_bloch_vector
#import matplotlib.pyplot as plt
```

```python
#Code
def add_k_fourier(circuit, k, n):
    """Aplica las rotaciones de fase necesarias para sumar k en la base de Fourier."""
    for j in range(n):
        angle = (2 * np.pi * k) / (2 ** (j + 1))
        circuit.p(angle, j)

def draper_adder(a, b, n):
    """
    Implementa el algoritmo de Draper para sumar dos números enteros a y b utilizando n qubits.
    Retorna el circuito cuántico listo para ser ejecutado.
    """
    qc = QuantumCircuit(n, n)  # n qubits y n bits clásicos para la medición

    # Paso 1: Inicialización del registro cuántico con el número a
    # Convertimos 'a' a binario y lo asignamos a los qubits
    a_bin = format(a, f'0{n}b')
    for i in range(n):
        if a_bin[n - i - 1] == '1':
            qc.x(i)

    # Paso 2: Aplicamos la Transformada de Fourier Cuántica (QFT)
    qc.append(QFT(num_qubits=n, do_swaps=False).to_gate(), range(n))

    # Paso 3: Aplicamos las rotaciones de fase para sumar 'b' en la base de Fourier
    add_k_fourier(qc, b, n)

    # Paso 4: Aplicamos la Transformada de Fourier Cuántica Inversa (QFT†)
    qc.append(QFT(num_qubits=n, do_swaps=False).inverse().to_gate(), range(n))

    # Paso 5: Medimos los qubits
    qc.measure(range(n), range(n))

    return qc
```

```python
# Ejemplo de uso
if __name__ == "__main__":
    # Números a sumar
    a = int(input("Introduce el primer número: "))#0
    b = int(input("Introduce el segundo número: "))#1
    n = 5  # Número de qubits (debe ser suficiente para representar 'a + b')

    # Verificación de que 'a' y 'b' pueden ser representados con 'n' qubits
    max_value = 2 ** n - 1
    if a > max_value or b > max_value:
        raise ValueError(f"Los números 'a' y 'b' deben ser menores o iguales a {max_value} para 'n' qubits.")

    # Creación y ejecución del circuito
    draper_qc = draper_adder(a, b, n)

    # Dibujamos el circuito
    #qc.draw('mpl')
    #import matplotlib.pyplot as plt
    #plt.show()

    # Ejecución del circuito en el simulador
    backend = Aer.get_backend('qasm_simulator')
    compiled_circuit = transpile(draper_qc,backend)
    job = backend.run(compiled_circuit, shots=1)
    result = job.result()
    counts = result.get_counts()

    # Procesamiento del resultado
    output = list(counts.keys())[0]
    decimal_result = int(output, 2)

    print(f"El resultado de sumar {a} y {b} es:")
    print(f"Representación binaria: {output}")
    #print(f"Representación binaria: {output[::-1]}")  # Invertimos el orden para mostrar correctamente
    print(f"Representación decimal: {decimal_result}")
```

# Introducción a noise models

Una de las funciones de qiskit-aer es la poder simular circuitos con ruido, pero antes de pasar a noise models, es necesario que hablemos de ruido.

## ¿Qué es el ruido?
Ruido, en el contexto de la computación cuántica, se refiere a cualquier fuente no deseada que altere el sistema cuántico que constituye nuestro hardware. El ruido es inevitable en el mundo físico, por lo que debemos diseñar e implementar algoritmos cuánticos que mantengan un alto desempeño incluso en presencia de ruido. El ruido que afecta las mediciones puede provenir de distintas fuentes, como errores en el hardware, ruido estocástico (aleatorio), errores de inicialización, entre otros. Cuando un sistema de hardware es afectado por el ruido, los qubits físicos corren el riesgo de perder su coherencia. ¿A qué nos referimos por coherencia? La coherencia es la propiedad de los estados cuánticos que permite mantener relaciones de fase entre los estados base, permitiendo la superposición. Si el sistema es afectado por ruido, los qubits perderán su coherencia y, por ende, su capacidad de estar en superposición, lo que limita su utilidad en cálculos cuánticos.

![](https://drive.google.com/uc?export=view&id=1TRCEnVZ2IDNxqZoTOIGMzrxak6RTjal4)

De manera general, podemos clasificar los tipos de ruido de la siguiente manera:

### Ruido coherente
Este tipo de ruido está asociado con errores sistemáticos que tendrán un efecto predecible en el sistema. Por sistemáticos nos referimos a errores que afectarán consistentemente a los qubits cada vez que una compuerta u operador es aplicado. El ruido coherente tiende a acumularse y empeora el estado de los qubits conforme el número de compuertas aumenta, provocando así resultados que difieren en gran medida de lo esperado.

### SPAM : State preparation and Measurement
Este tipo de errores están asociados a los procesos de inicialización y medición de qubits. En la inicialización, si un qubit debería ser preparado en el estado $|0\rangle$ pero, debido a imperfecciones en el proceso de preparación, se inicia en un estado mixto o en un estado incorrecto, esto introduce un error en el circuito. De manera similar, si al medir un qubit no se obtiene el estado verdadero en el que se encuentra, esto se debe a fallos en el aparato de medición, como el ruido de lectura o errores de calibración. Estos errores son descritos de manera probabilística y afectan la fidelidad del circuito, ya que impactan el estado inicial de los qubits o proporcionan un resultado de medición incorrecto.

### Ruido no-coherente
Finalmente, el error por ruido no-coherente es debido a la interacción del sistema físico de hardware con el medio ambiente. Dichas interacciones provocan que los qubits pierdan sus estados de coherencia de manera aleatoria. Contrario al ruido coherente, que introduce errores sistemáticos y predecibles, el ruido no-coherente provoca una degradación aleatoria e irreversible del estado cuántico del sistema. Algunos de los fenómenos que causan este tipo de errores incluyen decoherencia, fluctuaciones térmicas, emisión espontánea, y ruido de fase, todos los cuales actúan para aleatorizar los estados de los qubits y romper las correlaciones cuánticas esenciales para el procesamiento cuántico. Este tipo de errores lleva a la pérdida de información cuántica porque las interacciones son procesos irreversibles: una vez que la información se mezcla con el entorno, no se puede recuperar mediante operaciones reversibles.

Para una explicación más detallada e intuitiva de los tipos de error, invitamos a ver las dos pláticas del Dr. Zlatko Minev en la [Qiskit Global Summer School 2023](https://www.youtube.com/watch?v=3Ka11boCm1M&t=2200s).

Una vez descritos de manera muy general los tipos de errores por ruido que existen, procederemos a dar algunos ejemplos de implementaciones para los circuitos que hemos visto.

El paquete qiskit-aer contiene tres principales clases de ruido:
1. NoiseModel, el cual contiene modelos de ruidos para ser usados en simulación
2. QuantumError, el cual describe error de compuertas CPTP. Pueden ser aplicados antes de la medición.
3. ReadoutError el cual describe errores clásicos de lectura.

Al agregar un NoiseModel a un circuito, debemos escoger si queremos que actúe sobre todos los qubits o solamente a uno o unos en específico. Nos basaremos en el circuito de Bell para este ejemplo.

```python
"""
En este celda mostramos NoiseModels que afecten a todos los qubits
"""
# Crear un NoiseModel vacío
noise_model_1 = NoiseModel()

# Añadir depolarizing noise al Hadamard gate (1-qubit gate)
error_h = depolarizing_error(0.05, 1)  # 0.05 error rate for single-qubit gates
noise_model_1.add_all_qubit_quantum_error(error_h, ['h'])

# Añadir depolarizing noise al CNOT gate (2-qubit gate)
error_cx = depolarizing_error(0.05, 2)  # 0.05 error rate for two-qubit gates
noise_model_1.add_all_qubit_quantum_error(error_cx, ['cx'])

# Print noise model info
print(noise_model_1)
```
